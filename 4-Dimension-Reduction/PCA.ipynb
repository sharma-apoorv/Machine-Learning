## Power Iteration

We first implement the `power_iteration` method. Power Iteration is a Linear Algebra method for approximating the dominant Eigenvalues and Eigenvectors of a matrix.

```python
n, d = X.shape
```

```python
cov_matrix = 1 / (n - 1) * X.T.dot(X) # = np.cov(X.T)
```

```python
def eigenvalue(A, v):    
    return np.matmul(np.matmul(v.T, A), v)

def power_iteration(A, max_iters = 100, epsilon = 0.01):
    n, d = A.shape

    v = np.random.rand(n)
    ev = eigenvalue(A, v)
    
    i = 0
    while True:
        Av = A.dot(v)
        v_new = Av / np.linalg.norm(Av)
        ev_new = eigenvalue(A, v_new)
        
        # if np.abs(ev - ev_new) < epsilon: break
        
        v = v_new
        ev = ev_new
        
        i += 1
        if i > max_iters: break
    
    return ev_new, v_new
```

```python
w, v = power_iteration(cov_matrix)
```

```python
w_np, v_np = np.linalg.eig(cov_matrix)
```

We compare our methods to the numpy function, to ensure we have implemented the method correctly. 

```python
w, w_np[0]
```

```python
np.linalg.norm(v_np[:, 0] - v)
```

We then write a method to calculate the deflated covariance matrix. This function can be called each time we want to produce the next dominant eigenvalue and eigenvector

```python
def get_new_cov_matrix(M, ew, ev):
    return(M - ew * np.outer(ev, ev))
```

```python
cov_matrix_deflated = get_new_cov_matrix(cov_matrix, w, v)
w2, v2 = power_iteration(cov_matrix_deflated)
```

We again compare to the numpy implementation to ensure this method has been implemented correctly

## PCA

Here we generate all pairs of eigenvalues and eigenvectors using the method below. In each iteration we call `power_iteration`. This produces the dominant eigenpair. Then the covariance matrix is deflated. This process is repeated for all features in the input matrix

```python
def get_eigenpairs(A, max_iter = 10e2, epsilon = 0.01):
    # Initialization
    ev = [None] * A.shape[0]
    ew = [None] * A.shape[0]
    
    for i in range(0, A.shape[0]):
        ew[i], ev[i] = power_iteration(A, max_iter, epsilon)
        A = get_new_cov_matrix(A, ew[i], ev[i])

    return(np.array(ew), np.array(ev))
```

```python
A = 1 / (n - 1) * X.T.dot(X) # = np.cov(X.T)
w, v = get_eigenpairs(A)
```

```python
_n_components = 20
pca = PCA(n_components = _n_components)
```

We now project the train and test data onto the reduced dimensions. 

```python
X_transform = pca.fit_transform(X)
X_test_transform = pca.transform(X_test)
```

We now train a KNN model for the nearest neighbours. This is done for both the original dataset (before projection), as well as for the projected dataset. 

As a result, we can then compare the loss in accuracy due to the projection

```python
_n_neighbors = 1
neigh_original = KNeighborsClassifier(n_neighbors = _n_neighbors).fit(X, y)
neigh_reduced = KNeighborsClassifier(n_neighbors = _n_neighbors).fit(X_transform, y)
```

```python
orig_acc = neigh_original.score(X_test, y_test) * 100
reduce_acc = neigh_reduced.score(X_test_transform, y_test) * 100
```

```python
acc_change = (reduce_acc - orig_acc)/ orig_acc * 100

print(f'Original Accuracy: {orig_acc:.2f} %')
print(f'Transformed Accuracy: {reduce_acc:.2f} %')
print(f'\nThe accuracy changed by: {acc_change:.2f} %')
```

## Better `power_iteration` 

In the above implementaion the stopping criterion was once the loop reaches the max number of iterations. This may or may not always be the best criterion to use.

We could potentially add a better criterion that allows the eigenvector to converge faster. This would be as follows:

$$
abs(eigenvalue_{t-1} - eigenvalue_{t}) < \epsilon
$$
